```python
import pandas as pd 
import seaborn as sb 
import matplotlib.pyplot as plt
```

```python
data =  "./MICRODADOS_ENEM_2019_SAMPLE_43278.csv"  
dados = pd.read_csv(data)
```

```python
alunos_menor_14anos = dados.query("NU_IDADE <= 14") 
alunos_menor_14anos["SG_UF_RESIDENCIA"].value_counts().plot.pie(figsize=(10,8))
```

```python
plt.figure(figsize=(10,8)) 
sb.boxplot(x = "Q006",y="NU_NOTA_MT", data = dados)
plt.title("Boxplot das notas de matématica pela renda")
```

```python
renda_ordenada = dados["Q006"].unique() 
renda_ordenada.sort()
```

```python
plt.figure(figsize=(10,8)) 
sb.boxplot(x="Q006",y = "NU_NOTA_MT",data = dados, order = renda_ordenada) 
plt.title = ("Boxplot das notas de matemática pela renda")
```

```python
provas = ["NU_NOTA_CN","NU_NOTA_CH","NU_NOTA_MT","NU_NOTA_LC","NU_NOTA_REDACAO"] 
```

```python
dados[provas].sum()
```

```python
dados["NU_NOTA_TOTAL"] = dados[provas].sum(axis=1)  
dados.head()
```

```python
sb.displot(dados,x = "NU_NOTA_TOTAL")
```

```python
provas.append("NU_NOTA_TOTAL")
dados_com_nota_zero = dados[provas].query("NU_NOTA_TOTAL == 0") 
dados_com_nota_zero
```

```python
dados_sem_notas_zero = dados.query("NU_NOTA_TOTAL != 0") 
dados_sem_notas_zero.head()
```

```python
plt.figure(figsize=(10,6)) 
sb.boxplot(x="Q006", y = "NU_NOTA_TOTAL",data = dados_sem_notas_zero, order = renda_ordenada)
plt.title("Boxplot das notas de total pela renda")
```

```python
plt.figure(figsize=(10,6)) 
sb.boxplot(x="Q006", y = "NU_NOTA_TOTAL",data = dados_sem_notas_zero,hue ="IN_TREINEIRO" ,order = renda_ordenada)
plt.title("Boxplot das notas de total pela renda")
```
